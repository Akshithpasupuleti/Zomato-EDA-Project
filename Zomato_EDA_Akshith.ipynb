```python
df=pd.read_csv('zomato.csv',encoding='latin-1')
df.head()
```

```python
df.drop(['Country Code','Address','Locality','Locality Verbose','Currency','Switch to order menu','Longitude','Latitude','Switch to order menu'],axis=1)
```

```python
df.info()
```

```python
df=df.rename(columns={
    'Restaurant ID':'Restaurant_ID',
    'Restaurant Name':'Restaurant_Name',
    'Aggregate rating':'Aggregate_rating',
    'Votes':'Votes',
    'Price range':'Price_range' ,
    'Rating color':'Rating_color',
    'Rating text':'Rating_text'
})
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
top_cities=df['City'].value_counts().head(10) ## this will give you the head of top cities that appears
## most of the time
plt.figure(figsize=(10,6))
sns.barplot(x=top_cities.index,y=top_cities.values,palette='viridis')
plt.title("Top 10 Cities with the Most Restaurants")
plt.xticks(rotation=45)
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
from collections import Counter

cuisines_series = df['Cuisines'].dropna().apply(lambda x: x.split(',')[0])
top_cuisines = pd.Series(Counter(cuisines_series)).sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_cuisines.index, y=top_cuisines.values, palette='coolwarm')
plt.title("Top 10 Cuisines")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Average Cost for two', y='Aggregate_rating', data=df, hue='Price_range', palette='magma', alpha=0.6)
plt.title("Cost vs Rating")
plt.xlabel("Average Cost for Two")
plt.ylabel("Aggregate Rating")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Has Table booking', y='Aggregate_rating', data=df, palette='Set2')
plt.title("Rating by Table Booking Availability")
plt.xlabel("Has Table Booking")
plt.ylabel("Rating")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Aggregate_rating'], bins=20, kde=True, color='skyblue')
plt.title("Distribution of Restaurant Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
top_restaurants = df[['Restaurant_Name', 'Votes']].sort_values(by='Votes', ascending=False).drop_duplicates().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_restaurants['Restaurant_Name'], y=top_restaurants['Votes'], hue=top_restaurants['Restaurant_Name'], palette='plasma', legend=False)
plt.title("Top 10 Most Popular Restaurants by Votes")
plt.ylabel("Votes")
plt.xlabel("Restaurant")
plt.xticks(rotation=45)
plt.show()
```
